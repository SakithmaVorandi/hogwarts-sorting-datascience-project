```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/hogwarts_sorting.csv")
```

```python
df.drop(columns=['Dark Arts Knowledge','Quidditch Skills','Dueling Skills'], inplace=True)
```

```python
df.corr(numeric_only=True)
```

```python
sns.boxplot(x=df['Bravery'])
plt.title('Boxplot of Bravery')
plt.show()
plt.savefig('boxplot.png')
```

```python
sns.boxplot(x=df['Creativity'])
plt.title('Boxplot of creativity')
plt.show()
```

```python
sns.scatterplot(x=df['Bravery'], y=df['Ambition'], hue=df['House'])
plt.title('Scatter Plot of Bravery vs Ambition')
plt.show()
```

```python
sns.countplot(x='House', data=df)
plt.title('Count of Students in Each House')
plt.show()
```

```python
sns.pairplot(
    df,
    hue='House'
)

plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Bravery', hue='House', kde=True, multiple='stack')
plt.title('Distribution of Bravery Across Houses')
plt.show()
```
