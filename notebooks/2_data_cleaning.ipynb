```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'F:\DataScience Practical\hogwarts-sorting-datascience-project\data\raw\hogwarts_sorting.csv')
```

```python
df.isnull().sum() #checking for null values
```

```python
#dropping null values, inplace=True means the changes will be made to the original DataFrame
df.dropna(inplace=True) 

#filling null values in 'Bravery' column with the mean of the column
#Column is chosen if theres any null values, here it is already cleaned so chosed the Bravery for demonstration.
df['Bravery'].fillna(df['Bravery'].mean(), inplace=True)

##filling null values in 'Bravery' column with the mode of the column
df['Bravery'].fillna(df['Bravery'].mode()[0], inplace=True)


df.ffill(inplace=True)   # Forward Fill
df.bfill(inplace=True)   # Backward Fill
```

```python
df.duplicated().sum() # Count of duplicate rows

df.drop_duplicates(inplace=True) # Drop duplicate rows, inplace=True means the changes will be made to the original DataFrame
```

```python
# Dropping the 'Name' column from the DataFrame, inplace=True means the changes will be made to the original DataFrame
df.drop(columns=['Name'], inplace=True) 
```

```python
# Renaming columns for better readability and consistency
df.rename(columns={'Blood Status': 'Blood_Status'}, inplace=True) 
# Renaming columns for better readability and consistency
df.rename(columns={'Bravery': 'Bravery_Score'}, inplace=True)

# Creating a new column 'Blood_Status' by replacing spaces with underscores and converting to lowercase
df['Blood_Status']=df['Blood Status'].str.replace(' ', '_').lower()
```

```python
# Converting the 'House' column to a categorical data type for better memory usage and performance
df['House']=df['House'].astype('category')
```
