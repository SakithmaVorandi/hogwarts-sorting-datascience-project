```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'F:\DataScience Practical\hogwarts-sorting-datascience-project\data\raw\hogwarts_sorting.csv')
```

```python
df.corr(numeric_only=True) #Correlation matrix for numeric features
```

```python
#Boxplot for numeric features
sns.boxplot(data=df ['Bravery'])
plt.title('Boxplot of Bravery')
plt.savefig('boxplot.png')
plt.show()
```

```python
#Scatter plot for numeric features
sns.scatterplot(x=df['Bravery'], y=df['Ambition'], hue=df['House'])
plt.title('Ambition')
#plt.ylabel('Ambition')
plt.show()
```

```python
#Heatmap for correlation matrix
sns.countplot(x='House', data=df)
plt.title('Count of Students in each House')
plt.show()
```

```python
sns.pairplot(
    df,
    hue='House'
)

plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Bravery', hue='House', kde=True, multiple='stack')
plt.title('Distribution of Bravery Across Houses')
plt.show()
```
