```python
import pandas as pd
df=pd.read_csv("../data/raw/hogwarts_sorting.csv")

df.isnull().sum()
```

```python
df.dropna(inplace=True)
df['Bravery'].fillna(df['Bravery'].mean(), inplace=True)
df['Blood Status'].fillna(df['Blood Status'].mode()[0], inplace=True)

df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)

```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=['Name'], inplace=True)
```

```python
df.rename(columns={'Blood Status': 'Blood_Status'}, inplace=True)
df.rename(columns={'Bravery': 'Bravery_Score'}, inplace=True)

df['Blood_Status']=df['Blood Status'].str.replace(' ', '_').lower()
```

```python
df['House']=df['House'].astype('category')
```
